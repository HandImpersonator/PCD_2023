# Generate A matrix

```python
import numpy as np
import random
import time


def matrices():
    dim_a = input("Dimensión A: ")
    dim_b = input("Dimensión B: ")

    rows_a, cols_a = map(int, dim_a.split(","))
    rows_b, cols_b = map(int, dim_b.split(","))

    st1 = time.time()
    a = [[random.randint(0, 222) for _ in range(cols_a)] for __ in range(rows_a)]
    et1 = time.time() - st1

    st2 = time.time()
    b = np.random.randint(0, 222, (rows_b, cols_b))
    et2 = time.time() - st2

    return a, b, et1, et2


def multiplicacion_secuencial(a, b):
    rows_a = len(a)
    cols_a = len(a[0])
    cols_b = len(b[0])

    st = time.time()
    c = [[0 for _ in range(cols_b)] for __ in range(rows_a)]

    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                c[i][j] += a[i][k] * b[k][j]
    et3 = time.time() - st

    return c, et3
    

matrix_a, matrix_b, end_time_1, end_time_2 = matrices()

while len(matrix_a[0]) != len(matrix_b):
    print("Vuelve a generar las matrices, no son compatibles.")
    matrix_a, matrix_b, end_time_1, end_time_2 = matrices()
else:
    print("Se puede multiplicar.")
    matrix_c, end_time_3 = multiplicacion_secuencial(matrix_a, matrix_b)

print("Tiempo de ejecución sin numpy:", end_time_1, "segundos")
print("Tiempo de ejecución con numpy:", end_time_2, "segundos")
print("Tiempo de ejecución sin numpy:", end_time_3, "segundos")

```
