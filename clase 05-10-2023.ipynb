# Threads vs. process

## Threads ex.

```python
def do_nothing():
    while True:
        pass
```

```python
for i in range(20):
    th.Thread(target = do_nothing).start()

th.enumerate()
```

## Process Ex.

```python
for i in range(20):
    mp.Process(target = do_nothing()).start

mp.active_children()
```
